# Analysing Indeed Dataset:  Data Scientist/Analyst/Engineer


The data set was gotten from [Kaggle](https://www.kaggle.com/elroyggj/indeed-dataset-data-scientistanalystengineer)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline


data= pd.read_csv('indeed_job_dataset.csv')
data.head()
```

```python
data.count
```

```python
count_consult=data['Consulting and Business Services'].value_counts()[1]
count_internet=data['Internet and Software'].value_counts()[1]
count_bank=data['Banks and Financial Services'].value_counts()[1]
count_health=data['Health Care'].value_counts()[1]
count_insurance =data['Insurance'].value_counts()[1]
count_other_industries=data['Other_industries'].value_counts()[1]


various_industries=pd.DataFrame({"Num_of_Jobs": [count_consult,count_internet,count_bank,count_health,count_insurance,count_other_industries]}, 
                                index=['Consulting & Business Services','Internet & Software', 'Banks & Financial Services','Health Care','Insurance ','Other_industries'])
various_industries.plot.pie(y="Num_of_Jobs",figsize=(12,12),autopct="%1.1f%%")
```

```python
sns.set(style='whitegrid') # style can be 'darkgrid, whitegrid,white,dark,ticks'
plt.figure(figsize=(12,8))
ax=sns.countplot(x="No_of_Skills",data=data)
```

```python
python_count= data['python'].value_counts()[1]
r_count=data['r'].value_counts()[1]


languages=pd.DataFrame({'Programming Language':['Python','R'],'Demand':[python_count,r_count]})

ax=languages.plot.bar(x= 'Programming Language', y='Demand', figsize=(10,8),rot=0 , fontsize=16)
```

```python
data['Skill'].str.contains('Machine Learning').value_counts()
```

```python
mysql_count=data['Skill'].str.contains('MySQL').value_counts()[1]
oracle_count=data['Skill'].str.contains('Oracle').value_counts()[1]
db2_count=data['Skill'].str.contains('DB2').value_counts()[1]
access_count = data['Skill'].str.contains('Microsoft Access').value_counts()[1]
postgre_count = data['Skill'].str.contains('PostgreSQL').value_counts()[1]

rdbms_list=['MySQL','Oracle','DB2','Microsoft Access','PostgreSQL']
rdbms_demand_list=[mysql_count,oracle_count,db2_count,access_count,postgre_count]
rdbmsData=pd.DataFrame({'RDBMS':rdbms_list, 'Demand':rdbms_demand_list})
ax=rdbmsData.plot.bar(x='RDBMS', y='Demand', figsize=(12,8),fontsize=16, rot=0)
ax.set_ylabel('Job Demand Count', fontsize=10)
ax.set_xlabel("Relational Databases Management System", fontsize=10)
ax.set_title("Expected RDBMS Skill in Job Market for Data Science ", fontsize=20)
```

```python
power_bi_count=data['Skill'].str.contains('Power BI').value_counts()[1]
tableau_count=data['Skill'].str.contains('Tableau').value_counts()[1]
excel_count = data['Skill'].str.contains('Excel').value_counts()[1]



BI_tools=['Power BI','Tableau','Excel']
BI_tools_demand_list=[power_bi_count,tableau_count,excel_count ]
BI_tools_Data=pd.DataFrame({'BI Tools': BI_tools,'Demand':BI_tools_demand_list})

ax= BI_tools_Data.plot.bar(x='BI Tools',y='Demand', rot=0 , figsize=(10,8), fontsize=14)
```

```python
aws_count = data['Skill'].str.contains('AWS').value_counts()[1]
azure_count = data['Skill'].str.contains('Azure').value_counts()[1]
gcp_count = data['Skill'].str.contains('Google Cloud').value_counts()[1]


cloud_list = ['Amazon Web Service', 'Microsoft Azure', 'Google Cloud Platform']
cloud_demand_list = [aws_count, azure_count, gcp_count]
cloudDF = pd.DataFrame({'CloudService': cloud_list , 'Demand': cloud_demand_list})
ax = cloudDF.plot.bar(x='CloudService', y='Demand' , figsize=(14,8), rot=0, fontsize=16)
ax.set_ylabel("Job Demand Count", fontsize=16)
ax.set_xlabel("Cloud Service", fontsize=16)
ax.set_title("Expected Cloud Computing Skill in Data Science Job Market", fontsize=12)
```
